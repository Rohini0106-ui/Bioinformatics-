```python
#bar plot for gene expressrion
import matplotlib.pyplot as plt
import numpy as np
#sample data
condition = ['Control','Treatment1','Treatment2']
expression = [8.2, 7.3, 9.2]
#bar plot
plt.bar(condition, expression, color=['blue', 'pink', 'red'])
plt.xlabel('Condition')
plt.ylabel('Expression')
plt.title('Gene Expression Levels')
plt.show()
```

```python
#heatmap for gene expression
import seaborn as sns
import numpy as np

#sample data
genes = ['Gene1', 'Gene2', 'Gene3', 'Gene4']
conditions = ['Control', 'Treatment1', 'Treatment2']
data = np.random.rand(4, 3) * 10 # random gene data

#heatmap
sns.heatmap(data, annot=True, cmap='Blues', xticklabels=conditions, yticklabels=genes, )
plt.title('Gene Expression Heatmap')
plt.show()
```

```python
#valcano plot for differential expression
import numpy as np
import matplotlib.pyplot as plt

#sample data
np.random.seed(123)
logFC = np.random.normal(0, 2, 1000)
p_value = np.random.uniform(0.001, 0.1, 1000)
significant = (p_value < 0.05) & (np.abs(logFC) > 1)

#valcano plot
plt.scatter(logFC, -np.log10(p_value), c=significant, cmap='viridis', alpha=0.7)
plt.axhline( -np.log10(0.05), color='red', linestyle='--')
plt.axvline(1, color='red', linestyle='--')
plt.axvline(-1, color='red', linestyle='--')
plt.xlabel('log2(Fold Change)')
plt.ylabel('-log10(p-value)')
plt.title('Differential Expression Analysis')
plt.colorbar(label='Significant')
plt.show()
```

```python
#taxonomic abundance stacked bar plot
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#sample data
samples = ['Sample1', 'Sample2', 'Sample3']
taxa = ['Taxon1', 'Taxon2', 'Taxon3']
data = pd.DataFrame({
    'sample': samples * len(taxa),
    'Taxon': taxa * len(samples),
    'Abundance': np.random.randint(10, 60, len(samples) * len(taxa))
})

#piviot tble for stacked bar plot (use aggregation to handle duplication)
pivot_data = data.pivot_table(index='sample', columns='Taxon', values='Abundance', aggfunc='sum')

#stacked bar plot
pivot_data.plot(kind='bar', stacked=True, colormap="viridis")

#add lables and title
plt.title("XYZ")
plt.xlable("sample")
plt.ylable("Abundance")
plt.legend(title="Taxon")
plt.show()
```

```python
#stacked bar chhart
import numpy as np
import matplotlib.pyplot as plt

#data
samples = ['Sample1', 'Sample2', 'Sample3']
spe_a = [30,40,50]
spe_b = [60,80,10]
sep_c = [30,40,70]

#stacked bar chart
x = np.arange(len(samples))
plt.bar(x, spe_a, color='blue', label='spe_a')
plt.bar(x, spe_b, bottom=spe_a, color='red', label='spe_b')
plt.bar(x, sep_c, bottom=np.array(spe_a) + np.array(spe_b), lable='Spe_c', color='green')

plt.xticks(x, samples)
plt.xlabel('Samples')
plt.ylabel('Expression')
plt.title('Gene Expression Levels')
plt.legend()
plt.show()
```

```python
#grouped bar
import numpy as np
import matplotlib.pyplot as plt

departments =['ICU', 'Surgery', 'Ortho']
blaCTX_M = [60,40,30]
mecA = [50,35,20]
vanA = [40,25,15]
vfc = [30,40,50]

#grouped bar
x = np.arange(len(departments))

width = 0.15
plt.bar(x - 15 * width, blaCTX_M, color='blue', label='blaCTX_M')
plt.bar(x - 15 * width, mecA, color='red', label='mecA')
plt.bar(x * 15 * width, vanA, color='pink', label='vanA')
plt.bar(x * 15 * width, vfc, color='green', label='vfc')

plt.xticks(x, departments)
plt.xlabel('department')
plt.ylabel('Prevelence %')
plt.title('Prevelence by department')
plt.legend()
plt.show()
```

```python
#handling biological sequence
from Bio.Seq import Seq
my_seq = Seq("AGTACACTGGT")
reverse_seq = my_seq.reverse_complement()
print("Reverse Complement:", reverse_seq)
```

```python
#transcription and tranlation
rna_seq = my_seq.transcribe()
protein_seq = rna_seq.translate()
print("RNA Sequence:", rna_seq)
print("Protein Sequence:", protein_seq)
```

```python
#gc content calculation
gc_content = 100 * float(my_seq.count("G") +
my_seq.count("C")) / len(my_seq)
print("GC Content: %.2f%%" % gc_content)
```
